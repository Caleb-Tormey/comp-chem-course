## Part 3: The Morse Potential (Anharmonicity)

While the harmonic oscillator is a great approximation near equilibrium, real molecules can **dissociate** if given enough energy. A harmonic bond would require infinite energy to break!

To accurately model real bonds, we use the **Morse Potential**:
$$V_{\text{Morse}}(x) = D_e \left( 1 - e^{-a x} \right)^2$$

where:
* $D_e$ is the bond dissociation energy.
* $a = \sqrt{\frac{k}{2 D_e}}$ controls the width of the potential well.
* $x$ is the displacement from equilibrium.

Notice that for small $x$, $V_{\text{Morse}}(x) \approx \frac{1}{2} k x^2$, but for large positive $x$, $V_{\text{Morse}}(x) \to D_e$ (the bond breaks!).
