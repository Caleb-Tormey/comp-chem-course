## 1. Variables: The Computer's Memory

## 2. Lists: Grouping Data

Often, we have multiple related values, like the coordinates of an atom or the energies of different molecular states. Instead of making a hundred different variables, we put them in a **List**. 

Think of a list like a numbered rack of test tubes. 



**The Golden Rule of Computer Science:** Computers start counting at zero. The first item in your list is at index 0, the second is at index 1, and so on.

```python
# A list of atomic masses (Carbon, Nitrogen, Oxygen)
masses = [12.01, 14.01, 16.00]

# Getting the first item (Carbon)
print("First mass:", masses[0])

# Getting the last item (Oxygen)
print("Last mass:", masses[2])
```

## 3. Loops: The Workhorse of Simulation

In computational chemistry, we rarely do something once. We calculate the forces on an atom thousands of times per second. 

A `for` loop tells the computer to go through a collection of items (like a list) and perform the exact same action on every single one of them. 



[Image of a flowchart showing a for loop execution in Python]


Notice the **indentation**. In Python, spaces matter. Everything indented underneath the `for` statement is part of the loop.

```python
# Let's convert a list of energies from Joules to kJ
energies_J = [5000.0, 12000.0, 25000.0]

for energy in energies_J:
    energy_kJ = energy / 1000
    print(f"{energy} J is equal to {energy_kJ} kJ")
    
print("The loop has finished!") # Not indented, so it only prints once at the end.
```

## 4. Functions: Writing Your Own Formulas

In math, you define a function like $f(x) = x^2 + 2x$. In Python, you do the exact same thing using the `def` keyword. 

Functions take inputs (arguments), do some internal math, and explicitly `return` an output.

```python
# Define the function
def calculate_kinetic_energy(mass, velocity):
    ke = 0.5 * mass * velocity**2
    return ke

# Use the function
m = 2.0  # kg
v = 10.0 # m/s

my_energy = calculate_kinetic_energy(m, v)
print("Kinetic Energy:", my_energy, "Joules")
```

## Exercise: Plotting the Morse Potential

Now it's your turn! In computational chemistry, we often use the **Morse Potential** to model the energy of a diatomic molecule. The formula is:

$$V(r) = D_e [1 - e^{-a(r - r_e)}]^2$$

**Your Task:**
1. Use the interactive block below.
2. Define the parameters for a Hydrogen molecule ($H_2$): $D_e = 4.75$ eV, $r_e = 0.741$ Å, and $a = 1.93$ Å$^{-1}$.
3. Create a range of distances $r$ from 0.4 to 4.0 Å using `np.linspace`.
4. Calculate the potential $V$ and plot it.

```python
#| pyodide: true

import numpy as np
import matplotlib.pyplot as plt

# 1. Define the constants for H2
De = 4.75
re = 0.741
a  = 1.93

# 2. Create the distance array 'r'
r = np.linspace(0.4, 4.0, 100)

# 3. Calculate the Morse Potential 'V'
# Hint: use np.exp() for the exponential
V = De * (1 - np.exp(-a * (r - re)))**2

# 4. Create the plot
plt.figure(figsize=(8, 5))
plt.plot(r, V, color='teal', label='$H_2$ Morse Potential')
plt.axhline(De, color='gray', linestyle='--', label='Dissociation Limit')
plt.xlabel('Distance (Å)')
plt.ylabel('Energy (eV)')
plt.ylim(0, 6)
plt.legend()
plt.show()
```
